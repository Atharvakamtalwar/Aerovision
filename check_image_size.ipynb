```python
import os
from PIL import Image
from collections import Counter

def get_image_sizes(directory):
    # Dictionary to hold image size counts
    image_sizes = []

    # Walk through the directory and get image sizes
    for filename in os.listdir(directory):
        if filename.endswith(('.jpg')):
            filepath = os.path.join(directory, filename)
            try:
                with Image.open(filepath) as img:
                    image_sizes.append(img.size)
            except Exception as e:
                print(f"Error opening {filepath}: {e}")
                continue

    # Count unique image sizes
    size_counts = Counter(image_sizes)
    
    # Sort image sizes by count in descending order
    sorted_sizes = sorted(size_counts.items(), key=lambda x: x[1], reverse=True)

    return sorted_sizes

# Example usage
directory_path = r"D:\Military Aircraft Detection Dataset(1)\images\train"  # Replace with your directory path
image_size_counts = get_image_sizes(directory_path)

print("Unique image sizes and their counts (sorted in descending order):")
for size, count in image_size_counts:
    print(f"Size: {size}, Count: {count}")
```
